```python
# Creating two new columns for card catgory with attried and existing customers

existing_customers_by_card = df[df['Attrition_Flag'] == 'Existing Customer'].groupby('Card_Category')['Total_Trans_Amt'].sum()
attrited_customers_by_card = df[df['Attrition_Flag'] == 'Attrited Customer'].groupby('Card_Category')['Total_Trans_Amt'].sum()

summary = pd.DataFrame({
    'Existing_Customers_Total': existing_customers_by_card,
    'Attrited_Customers_Total': attrited_customers_by_card
}).reset_index()

print(summary)
```

```python
df.columns
```

```python
# Create a new dataframe with row-level data for correlation analysis
correlation_df = df[['Total_Revolving_Bal', 
                     'Customer_Age', 
                     'Total_Ct_Chng_Q4_Q1', 
                     'Avg_Utilization_Ratio', 
                     'Avg_Open_To_Buy', 
                     'Months_on_book']]

# Add the newly created columns (row-level data)
correlation_df['Existing_Customers_by_Card'] = df['Card_Category'].map(existing_customers_by_card)
correlation_df['Attrited_Customers_by_Card'] = df['Card_Category'].map(attrited_customers_by_card)

# Calculate the correlation matrix
correlation_matrix = correlation_df.corr()
print(correlation_matrix)


```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix for Card Categories ")
plt.show()
```

```python
# Grouping by Card_Category and Attrition_Flag
card_category_analysis = df.groupby(['Card_Category', 'Attrition_Flag']).agg(
    Total_Transactions=('Total_Trans_Amt', 'sum'),
    Average_Balance=('Total_Revolving_Bal', 'mean'),
    Average_Transactions=('Total_Trans_Ct', 'mean'),
    Average_Utilization_Ratio=('Avg_Utilization_Ratio', 'mean'),
    Average_Open_To_Buy=('Avg_Open_To_Buy', 'mean'),
    Average_months_on_book=('Months_on_book', 'mean'),
).reset_index()

print(card_category_analysis)
```

```python
# Now lets see how avg utilization ratio correlates with total revolving balance by attrition flag
plt.figure(figsize=(10,6))
sns.scatterplot(x='Avg_Utilization_Ratio', y='Total_Revolving_Bal', data=df, hue='Attrition_Flag')
plt.title('Scatterplot of Avg Utilization Ratio vs Total Revolving Balance by Attrition Flag')
plt.xlabel('Avg Utilization Ratio')
plt.ylabel('Total Revolving Balance')
plt.show()
```

```python
# visualize attrition rates for customers with months on book with a bar chart
df['months_bucket'] = pd.cut(df['Months_on_book'], bins=[0, 24, 48, 72, 96, 120, 144, 168, 192, 216, 240, 264, 288, 312, 336, 360], labels=["0-24", "25-48", "49-72", "73-96", "97-120", "121-144", "145-168", "169-192", "193-216", "217-240", "241-264", "265-288", "289-312", "313-336", "337-360"])
attrition_rates = df.groupby('months_bucket')['Attrition_Flag'].value_counts(normalize=True).unstack()
attrition_rates.plot(kind='bar', stacked=True)
plt.title('Attrition Rates by Months on Book')
plt.xlabel('Months on Book')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
#  so now lets make heatmap to see the correlation for avg utilization ratio by card category and attrition flag
heatmap_category = df.pivot_table( index="Card_Category", columns="Attrition_Flag",values="Avg_Utilization_Ratio", aggfunc="mean")

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_category,annot=True,cmap="coolwarm",fmt=".2f",cbar_kws={"label": "Avg Utilization Ratio"})
plt.title("Avg Utilization Ratio by Card Category and Attrition Flag", fontsize=14)
plt.xlabel("Attrition Flag", fontsize=12)
plt.ylabel("Card Category", fontsize=12)
plt.show()
```

# as we saw in the last three charts :
# - The number of attried customers have high utilization ratio which indicates financial stress
# - The clients with less months on book leaves which indicates non satsifaction, so here is an issue for grabbing new clients.
# - Blue card holders have higher utilization values but lower engagement which indicates they can leave the bank soon.

# I see here we should focus on clients with less months on book and especially who have blue card category.
